```python
def drop_recovery_difference(filename: str, drop: np.ndarray, recovery: np.ndarray):
    if filename in ['1c9db047-e335-46ac-8039-effd8589b25b', '1cbce6e5-9f0b-419f-9527-7add4e255217']:
        pass
        return
    data = pd.read_csv(filename, index_col=None, sep='\\s+')
    print('/' * 15 + filename + '/' * 15)
    

    if drop.size == 2:
        drop_min, drop_max = drop.reshape(2)
        if drop_max != 0:  
            print("-" * 10)
            print("Drop percentage difference: ", (1 - (drop_min / drop_max)) * 100)
            print("-" * 10)

    if recovery.size == 2:
        recovery_min, recovery_max = recovery.reshape(2)
        if recovery_min != 0:  
            print("-" * 10)
            print("Recovery percentage difference: ", ((recovery_max / recovery_min) - 1) * 100)
            print("-" * 10)

    if drop.size > 2:
        for drop_min, drop_max in drop:
            if drop_max != 0:
                print("-" * 10)
                print("Drop percentage difference: ", (1 - (drop_min / drop_max)) * 100)
                print("-" * 10)
            else:
                print("Error: Division by zero in drop calculation.")

    if recovery.size > 2:
        for recovery_min, recovery_max in recovery:
            if recovery_min != 0:
                print("-" * 10)
                print("Recovery percentage difference: ", ((recovery_max / recovery_min) - 1) * 100)
                print("-" * 10)

    print('/' * 66)
```

```python
labeled_df = pd.read_csv('../raw_data/ground_truth.csv')

for filename in labeled_df['file']:
    drop = labeled_df[labeled_df['file'] == filename]['drop'].item()
    recovery = labeled_df[labeled_df['file'] == filename]['recovery'].item()

    if drop != '[]':
        drop = np.array(ast.literal_eval(drop)).astype(float)
    else:
        continue

    if recovery != '[]':
        recovery = np.array(ast.literal_eval(recovery)).astype(float)
    else:
        continue
    
    if drop.size == 0 and recovery.size == 0:
        continue

    drop_recovery_difference(filename, drop, recovery)
```
