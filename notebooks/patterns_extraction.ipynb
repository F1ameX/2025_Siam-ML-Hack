```python
labeled_df = pd.read_csv('../raw_data/ground_truth.csv')
for filename in labeled_df['file']:
    drop = labeled_df[labeled_df['file'] == filename]['drop'].item()
    recovery = labeled_df[labeled_df['file'] == filename]['recovery'].item()
    if drop != '[]':
        drop = ast.literal_eval(drop)
    if recovery != '[]':
        recovery = ast.literal_eval(recovery)
        
    if len(drop) == 0 and len(recovery) == 0:
        continue
    drop_recovery_difference(filename, drop, recovery)  
```
